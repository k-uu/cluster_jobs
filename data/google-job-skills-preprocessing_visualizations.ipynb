```python
# Classical preprocessing: 'Minimum Qualifications', 'Preferred Qualifications'
col = ['Title', 'Minimum Qualifications', 'Preferred Qualifications']
punc = '''!()-[]{};:'"\,<>./?@#$%^&*_~'''

for ele in col:
    # lowercasing
    df[ele + '-Operated'] = df[ele].str.lower()
    # remove \n
    df[ele + '-Operated'] = df[ele + '-Operated'].replace('\n', ' ', regex=True)
    # remove punctuation using regex
    df[ele + '-Operated'] = df[ele + '-Operated'].replace(f"[{punc}]", "", regex=True)

df
```

```python
# removing stopwords

import nltk
nltk.download('stopwords')

from nltk.corpus import stopwords

# Create a set of stop words 
stop_words = set(stopwords.words('english')) 

df['Minimum Qualifications-Operated'] = df['Minimum Qualifications-Operated'].astype(str).apply(lambda x: ' '.join([word for word in x.split() if word not in (stop_words)]))
df['Preferred Qualifications-Operated'] = df['Preferred Qualifications-Operated'].astype(str).apply(lambda x: ' '.join([word for word in x.split() if word not in (stop_words)]))
df
```

# Choosing a representative description by calculating TF-IDF score for each role

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
roles = sorted(df['Category'].value_counts().index.tolist())
roles
```

```python
# Drop rows with 'nan'
df_drop = df[df['Preferred Qualifications-Operated'] != 'nan']
df_drop.describe()
# 1250 records to 1236 records
```

```python
roles_desc_unoperated = {}

for role in roles:
    roles_desc_unoperated[role] = df_drop[df_drop['Category'] == role]['Preferred Qualifications'].to_list()

print(len(roles))
# roles_desc_unoperated['Hardware Engineering']
```

```python
roles_desc = {}

for role in roles:
    roles_desc[role] = df_drop[df_drop['Category'] == role]['Preferred Qualifications-Operated'].to_list()

print(len(roles))
# roles_desc
```

```python
vectorizer = TfidfVectorizer()

roles_rep = {}

for role in roles:
    text = roles_desc[role]
    true_text = roles_desc_unoperated[role]
    X = vectorizer.fit_transform(text)
    terms = vectorizer.get_feature_names_out()
    
    # print(terms)
    # print(X.shape)

    tfidf = pd.DataFrame(X.toarray(), columns=terms)
    # print(tfidf)
    
    tfidf = tfidf.replace(0, np.NaN)
    tfidf['score'] = tfidf.mean(axis=1, skipna=True)
        
    rep_index = tfidf['score'].idxmax()
    print(rep_index)
    # print(text[rep_index])
    roles_rep[role] = [true_text[rep_index]]

# roles_rep
# tfidf['accelerator'].value_counts(dropna=False)
# tfidf
```

```python
pd.set_option('display.max_colwidth', None)
```

```python
df_rep = pd.DataFrame.from_dict(roles_rep, orient='index', columns=['Representative Description']).reset_index()
df_rep = df_rep[df_rep['index']!='Administrative']
df_rep.rename(columns={'index': 'Role'}, inplace=True)
df_rep
```

We concluded that perplexity=10 is the best for our visualization.

```python
from sklearn.manifold import TSNE

def tsne(embedding_df, PERP, RANDOM_STATE, n_iter):
    tsne = TSNE(n_components=3, perplexity=PERP, random_state=RANDOM_STATE, n_iter=n_iter)
    X_tsne = tsne.fit_transform(embedding_df)
    return X_tsne
```

```python
from time import time
import matplotlib.pyplot as plt
import plotly.io as pio

perplexities = [5, 10, 30, 50]
RANDOM_STATE = 42

for i, perplexity in enumerate(perplexities):

    t0 = time()

    X_tsne = tsne(embedding_df, perplexity, RANDOM_STATE, n_iter=1000)
    t1 = time()
    print("uniform grid, perplexity=%d in %.2g sec" % (perplexity, t1 - t0))
    
    fig = px.scatter_3d(
        X_tsne, x=0, y=1, z=2,
        color=roles_text)
    fig.update_traces(marker_size=2)
    
    fig.update_layout(
        title="t-SNE Visualization of Representative Description Perplexity=%d" % perplexity,
        scene=dict(
            xaxis_title="First t-SNE",
            yaxis_title="Second t-SNE",
            zaxis_title="Third t-SNE"
        )
    )
    
    pio.write_html(fig, 'tSNE_plot_perplexity_%d.html' % perplexity)
    fig.show()
```

# MDS visualization

```python
# Calculate cosine distance from cosine similarity
from sklearn.metrics.pairwise import cosine_similarity
similarity_matrix = cosine_similarity(embedding_df) # embeddings)
# cosine similarity takes value in the range [-1, 1] with 1 being the most similar
distance_matrix = 1 - similarity_matrix
# cosine distance
distance_matrix
```

```python
from sklearn.manifold import MDS
from sklearn.cluster import KMeans

# Reduction objects
manifold3D = MDS(n_components=3, dissimilarity='precomputed', random_state=42)
manifold2D = MDS(n_components=2, dissimilarity='precomputed', random_state=42)

# compressed data in 2 and 3 dimensions
manifold_data3D = manifold3D.fit_transform(distance_matrix)
manifold_data2D = manifold2D.fit_transform(distance_matrix)

# creates a dataframe to visualize each reduced version
manifold3D_df = pd.DataFrame(data =  manifold_data3D, 
                            columns = ['component_1', 'component_2', 'component_3'])
manifold2D_df = pd.DataFrame(data =  manifold_data2D, 
                            columns = ['component_1', 'component_2'])
```
